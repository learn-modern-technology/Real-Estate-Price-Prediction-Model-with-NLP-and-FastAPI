## Data Pre-processing

### I. Splitting the Location into City, State and Country

```python
df['Location'].unique()
```

```python
df['City'] = df['Location'].apply(lambda x : x.split(',')[0].strip())
df['State'] = df['Location'].apply(lambda x : x.split(',')[1].strip())
df['Country'] = df['Location'].apply(lambda x : x.split(',')[2].strip())
df.head()
```

### II. Cleaning Text Features based on its values

#### A. Property Type

```python
df['Propert Type'].value_counts()
```

```python
import re
regx_numbers = re.compile(r"[-+]?(\d*\.\d+|\d+)")
df['PropertyTypeCleaned'] = df['Propert Type'].apply(lambda x : regx_numbers.findall(x)[0] 
                                                     if len(regx_numbers.findall(x)) > 0 else 0)
df['PropertyTypeCleaned'].value_counts()
```

#### B. Sub-Area

```python
df['Sub-Area'].value_counts()
```

```python
len(df['Sub-Area'].value_counts())
```

```python
df['SubAreaCleaned'] = df['Sub-Area'].apply(lambda x: x.capitalize().strip())
df.head()
```

```python
df['SubAreaCleaned'].nunique()
```

#### C. Company Name

```python
df['Company Name'].nunique()
```

```python
df['Company Name'].unique()
```

```python
df['CompanyNameCleaned'] = df['Company Name'].apply(lambda x : x.capitalize())
df.head()
```

```python
df['CompanyNameCleaned'].nunique()
```

#### D. TownShip Name/ Society Name

```python
df['TownShip Name/ Society Name'].nunique()
```

```python
df['TownShip Name/ Society Name'].value_counts()
```

```python
df['TownShipSocietyNameCleaned'] = df['TownShip Name/ Society Name'].apply(lambda x: x.capitalize())
df.head()
```

```python
df['TownShipSocietyNameCleaned'].nunique()
```

#### E. Description

```python
df['DescriptionCleaned'] = df['Description'].apply(lambda x : x.capitalize())
df.head()
```

### III. Encoding Categorical Features

```python
df.info()
```

```python
df['ClubHouseEncoded'] = df['ClubHouse'].apply(lambda x: x.lower().strip()).map({'yes': 1, 'no': 0})
df['School/UniversityInTownshipEncoded'] = df['School / University in Township '].apply(lambda x: x.lower().strip()).map({'yes': 1, 'no': 0})
df['HospitalInTownShipEncoded'] = df['Hospital in TownShip'].apply(lambda x: x.lower().strip()).map({'yes': 1, 'no': 0})
df['MallInTownShipEncoded'] = df['Mall in TownShip'].apply(lambda x: x.lower().strip()).map({'yes': 1, 'no': 0})
df['ParkJoggingTrackEncoded'] = df['Park / Jogging track'].apply(lambda x: x.lower().strip()).map({'yes':1, 'no':0})
df['SwimmingPoolEncoded'] = df['Swimming Pool'].apply(lambda x: x.lower().strip()).map({'yes':1, 'no':0})
df['GymEncoded'] = df['Gym'].apply(lambda x: x.lower().strip()).map({'yes':1, 'no':0})
```

```python
df.head()
```

### IV. Cleaning Continuous Features and Missing Values

#### A. Property Area in Sq. Ft.

```python
df['Property Area in Sq. Ft.'].unique()
```

```python
def get_average_area(x):
    x = regx_numbers.findall(x)
    if len(x) == 1:
        return (np.float(x[0]))
    elif len(x) == 2:
        return (np.float(x[0]) + np.float(x[1]))/2
    else:
        return -1
```

```python
import re
regx_numbers = re.compile(r"[-+]?(\d*\.\d+|\d+)")
df['PropertyAreainSqFtCleaned'] = df['Property Area in Sq. Ft.'].apply(lambda x : get_average_area(str(x)))
df.head()
```

```python
df['PropertyAreainSqFtCleaned'].unique()
```

#### A. Price in lakhs

```python
df['Price in lakhs'].unique()
```

```python
import re
regx_numbers = re.compile(r"[-+]?(\d*\.\d+|\d+)")
df['PriceInLakhsCleaned'] = df['Price in lakhs'].apply(lambda x: np.float(regx_numbers.findall(str(x))[0]) 
                                                  if len(regx_numbers.findall(str(x)))>0 else np.NaN)
df.head()
```
