## III. Data Pre-processing

```python
df_initial = df.copy
modified_df = pd.DataFrame()
```

### I. Splitting the Location into City, State and Country

```python
df['Location'].unique()
```

```python
modified_df['City'] = df['Location'].apply(lambda x : x.split(',')[0].strip())
modified_df['State'] = df['Location'].apply(lambda x : x.split(',')[1].strip())
modified_df['Country'] = df['Location'].apply(lambda x : x.split(',')[2].strip())
modified_df.head()
```

### II. Cleaning Text Features based on its values

#### A. Property Type

```python
df['Propert Type'].nunique()
```

```python
import re
regx_numbers = re.compile(r"[-+]?(\d*\.\d+|\d+)")
modified_df['PropertyType'] = df['Propert Type'].apply(lambda x : regx_numbers.findall(x)[0] 
                                                     if len(regx_numbers.findall(x)) > 0 else 0)
modified_df['PropertyType'].nunique()
```

#### B. Sub-Area

```python
df['Sub-Area'].value_counts()[0:10]
```

```python
len(df['Sub-Area'].value_counts())
```

```python
modified_df['SubArea'] = df['Sub-Area'].apply(lambda x: x.capitalize().strip())
df.head()
```

```python
modified_df['SubArea'].nunique()
```

#### C. Company Name

```python
df['Company Name'].nunique()
```

```python
df['Company Name'].unique()
```

```python
modified_df['CompanyName'] = df['Company Name'].apply(lambda x : x.capitalize())
modified_df.head()
```

```python
modified_df['CompanyName'].nunique()
```

#### D. TownShip Name/ Society Name

```python
df['TownShip Name/ Society Name'].nunique()
```

```python
df['TownShip Name/ Society Name'].value_counts()[0:10]
```

```python
modified_df['TownshipSocietyName'] = df['TownShip Name/ Society Name'].apply(lambda x: x.capitalize())
df.head()
```

```python
modified_df['TownshipSocietyName'].nunique()
```

#### E. Description

```python
modified_df['Description'] = df['Description'].apply(lambda x : x.capitalize())
modified_df.head()
```

### IV. Cleaning Continuous Features and Missing Values

#### A. Property Area in Sq. Ft.

```python
df['Property Area in Sq. Ft.'].unique()
```

```python
def get_average_area(x):
    x = regx_numbers.findall(x)
    if len(x) == 1:
        return (np.float(x[0]))
    elif len(x) == 2:
        return (np.float(x[0]) + np.float(x[1]))/2
    else:
        return -1
```

```python
import re
regx_numbers = re.compile(r"[-+]?(\d*\.\d+|\d+)")
modified_df['PropertyAreainSqFt'] = df['Property Area in Sq. Ft.'].apply(lambda x : get_average_area(str(x)))
modified_df.head()
```

```python
modified_df['PropertyAreainSqFt'].unique()
```

#### A. Price in lakhs

```python
df['Price in lakhs'].unique()
```

```python
import re
regx_numbers = re.compile(r"[-+]?(\d*\.\d+|\d+)")
modified_df['PriceInLakhs'] = df['Price in lakhs'].apply(lambda x: np.float(regx_numbers.findall(str(x))[0]) 
                                                  if len(regx_numbers.findall(str(x)))>0 else np.NaN)
modified_df.head()
```

```python
modified_df['ClubHouse'] = df['ClubHouse'].apply(lambda x: x.lower().strip())
modified_df['School/UniversityInTownship'] = df['School / University in Township '].apply(lambda x: x.lower().strip())
modified_df['HospitalInTownShip'] = df['Hospital in TownShip'].apply(lambda x: x.lower().strip())
modified_df['MallInTownShip'] = df['Mall in TownShip'].apply(lambda x: x.lower().strip())
modified_df['ParkJoggingTrack'] = df['Park / Jogging track'].apply(lambda x: x.lower().strip())
modified_df['SwimmingPool'] = df['Swimming Pool'].apply(lambda x: x.lower().strip())
modified_df['Gym'] = df['Gym'].apply(lambda x: x.lower().strip())
```
